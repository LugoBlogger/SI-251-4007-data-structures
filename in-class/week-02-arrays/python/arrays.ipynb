Dot product of two vectors (of the same length) is the sum of the products
of their corresponding components

```python
x = [0.30, 0.60, 0.10]
y = [0.50, 0.10, 0.40]

total = 0.0
n = len(x)
for i in range(n):
  total += x[i]*y[i]

total
```

Trace of dot product computation

| `i` | `x[i]` | `y[i]` | `x[i]*y[i]` | `total` |
|-----|--------|--------|-------------|---------|
|     |        |        |             | `0.00`  |
| `0` | `0.30` | `0.50` | `0.15`      | `0.15`  |
| `1` | `0.60` | `0.10` | `0.06`      | `0.21`  |
| `2` | `0.10` | `0.40` | `0.04`      | `0.25`  |

- Mutability

```python
# reverse the order of the elements in an array `a[]`

a = [3, 1, 4, 1, 5, 9, 2]
n = len(a)
for i in range(n // 2):
  temp = a[i]
  a[i] = a[n-1-i]
  a[n-1-i] = temp

a
```

Trace for reversing the array `a`

- Iteration

```python
# Calculate the average of a using range(len(a)) and iterate by calling
# the element of a by its index
a = [3, 1, 4, 1, 5, 9, 2]

total = 0.0
for i in range(len(a)):
  total += a[i]

average = total / len(a)
average
```

```python
# Calcualte the average by directly accessing the element of a
a = [3, 1, 4, 1, 5, 9, 2]

total = 0.0
for v in a:
  total += v

average = total / len(a)
average
```

### Array aliases and copies

- Aliasing

```python
x = [0.30, 0.60, 0.10]
y = x
x[1] = 0.99
y
```

- Copying and slicing

```python
x = [0.30, 0.60, 0.10]
y = []
for v in x:
  y += [v]

x[1] = 0.99
y
```

```python
# A shorter way to copy an array in Python
x = [0.30, 0.60, 0.10]
y = x[:]

x[1] = 0.99
y
```

### Sample applications of arrays

- Representing playing cards

```python
SUITS = ['Clubs', 'Diamonds', 'Hearts', 'Spades']
RANKS = ['2', '3', '4', '5', '6', '7', '8', '9', '10', 'Jack', 'Queen', 'King', 'Ace']

seed = 260307
rng = np.random.default_rng(seed)

# draw a card
rank = rng.integers(0, len(RANKS))
suit = rng.integers(0, len(SUITS))
print(f"{RANKS[rank]} of {SUITS[suit]}")
```

```python
# initialize 52 cards and take randomly from it
deck = []
for rank in RANKS:
  for suit in SUITS:
    card = rank + ' of ' + suit
    deck += [card]

# display(deck)

card_idx = rng.integers(0, len(deck))
print(deck[card_idx])
```
