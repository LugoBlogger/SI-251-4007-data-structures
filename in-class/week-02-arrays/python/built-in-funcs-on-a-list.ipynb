
aggregation and analysis

```python
a = [0.3, 0.6, 0.1, 0.4, 0.5, -0.1]
print(len(a))
print(sum(a))
print(min(a))
print(max(a))

a = [True, True, False]
print(any(a))
print(all(a))

a = [False, False, False]
print(any(a))

a = [True, True, True]
print(all(a))
```

transformation and iteration

```python
a = [0.3, 0.6, 0.1, 0.4, 0.5, -0.1]
print(sorted(a))
print(list(reversed(a)))
print(list(enumerate(a)))


a = [ 0.3, 0.6, 0.1, 0.4, 0.5, -0.1]
b = [-0.5, 0.2, 1.2, 2.3, 0.2, -0.3]
print(list(zip(a, b)))

print(list(map(str, a)))

print(list(filter(lambda x: x > 0, a)))
```

construction and conversion

```python
a = [-3, 4, 5, -1, 8, 2]
print(list(a))

list(iter(a))

iter_a = iter(a)
for _ in range(len(a)):
  print(next(iter_a), end=", ")
```

inspection and utilities

```python
a = [-3, 4, 5, -1, 8, 2]
print(type(a))
print(id(a))
print(repr(a))
print(str(a))
print(dir(a))
print(help(a))
```
