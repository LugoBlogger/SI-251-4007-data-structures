An _abstraction_ is a simplified description of something that captures
its essential elements while supressing all other details.    

An _bject-oriented programming_, a style of programming that integrates code and data.
In object-oriented programming, we break a large and potentially complex 
program into a set of interacting elements, or _objects_.

This idea originates from modeling (in-software) real-world entities such as
electrons, people, building, or solar systems and readily extend to modeling
abstract entities such as bits, numbers, color, images, or programs.

The ability to define a new data types and to manipulate objects holding 
data type values is also known as _data abstraction_, and lead us to a style
of modular programming that naturally extends the _function abstraction_ style

A data type allows us to isolate _data_ as well as functions.

Our mantra for this discussion
> Whenever you can clearly separate data and associated tasks within a computation, 
> you should do so.

### Methods

A method is a function associated with a specified object (and, by extension, 
with the type of that object).

We can _call_ (or _invoke_) a method by using a variable name, followed by 
the _dot operator_ (`.`) followed by the method name, followed by a list
of arguments delimited by commas and enclosed in parentheses

```python
# Determine the number of bits in the binary representation of an `int` value
x = 3**100
bits = x.bit_length()
print(bits)
```

### Applications of string processing: genomics

```python
def is_potential_gene(dna):
  # number of bases is multiple of 3
  if (len(dna) % 3) != 0: return False

  # starts with start codon
  if not dna.startswith("ATG"): return False

  # no intervening stop codons
  for i in range(len(dna) - 3):
    if i % 3 == 0:
      if dna[i:i+3] == "TAA": return False
      if dna[i:i+3] == "TAG": return False
      if dna[i:i+3] == "TGA": return False

  # ends with a stop codon
  if dna.endswith("TAA"): return True
  if dna.endswith("TAG"): return True
  if dna.endswith("TGA"): return True

  return False
```

```python
is_potential_gene("ATGCGCCTGCGTCTGTACTAG")
```

```python
is_potential_gene("ATGCGCTGCGTCTGTACTAG")
```
