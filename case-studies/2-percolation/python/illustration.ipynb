```python
import numpy as np
import matplotlib.pyplot as plt

from matplotlib import colors
```

```python
plt.rcParams.update(plt.rcParamsDefault)
plt.rcParams.update({
  'font.size': 16, 
  'grid.alpha': 0.25})
```

```python
percolate_data = np.array(
  [[2, 2, 1, 1, 1, 2, 2, 2],
   [0, 2, 2, 1, 1, 1, 1, 1],
   [0, 0, 0, 2, 2, 1, 1, 2],
   [2, 2, 0, 0, 2, 1, 1, 1],
   [2, 0, 0, 0, 2, 1, 1, 2],
   [2, 0, 2, 2, 2, 2, 1, 1],
   [0, 2, 0, 2, 1, 1, 1, 1],
   [0, 0, 0, 0, 2, 1, 0, 0]])

non_percolate_data = np.array(
  [[2, 2, 1, 1, 2, 1, 2, 2],
   [1, 1, 1, 1, 1, 2, 2, 2],
   [1, 1, 2, 2, 1, 1, 2, 2], 
   [2, 2, 1, 1, 1, 1, 1, 2],
   [0, 2, 2, 2, 2, 2, 1, 1],
   [2, 0, 2, 0, 0, 0, 2, 2],
   [2, 0, 0, 2, 0, 0, 2, 0],
   [0, 2, 0, 2, 2, 2, 0, 2]
])
```

```python
# dpi at least 200 to align gridlines to the imshow boxes
fig, axes = plt.subplots(nrows=1, ncols=2, sharey=True, dpi=200)

# create discrete colormap
cmap = colors.ListedColormap(["w", "tab:blue", "k"])
bounds = [0, 1, 2, 3]
norm = colors.BoundaryNorm(bounds, cmap.N)

n = len(percolate_data)
axes[0].imshow(percolate_data, cmap=cmap, norm=norm, extent=[0, n, 0, n])

axes[0].grid(which="major", axis="both", linestyle="-", color="k", linewidth=1)
axes[0].set_xticks(np.arange(n+1))
axes[0].set_yticks(np.arange(n+1))
axes[0].tick_params(axis="x", which="major", bottom=False, labelbottom=False)
axes[0].tick_params(axis="y", which="major", left=False, labelleft=False)

n = len(non_percolate_data)
axes[1].imshow(non_percolate_data, cmap=cmap, norm=norm, extent=[0, n, 0, n])

axes[1].grid(which="major", axis="both", linestyle="-", color="k", linewidth=1)
axes[1].set_xticks(np.arange(n+1))
axes[1].set_yticks(np.arange(n+1))
axes[1].tick_params(axis="x", which="major", bottom=False, labelbottom=False)
axes[1].tick_params(axis="y", which="major", left=False, labelleft=False)

plt.show(fig)
```
